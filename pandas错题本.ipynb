```python
frames = [df_apple, df_tsla, df_ibm, df_msft]
df = pd.concat(frames)
df.set_index(keys='stock', append=True, inplace=True) 
#or df.set_index('stock', append=True, inplace=True) 
#指定要把 DataFrame 中的 stock 列 设置为新的索引列。append=True 保留原有索引，将新索引追加到末尾（形成多级索引）。
#如果是append=False（默认）：删除原有索引，用 stock 列替换成新索引；
vol = df['Volume']
vol = pd.DataFrame(vol)

date = vol.index.get_level_values(0) #从多级索引中提取第0级（日期索引）
date = pd.DatetimeIndex(date) #确保日期索引是 DatetimeIndex 类型（避免普通字符串/数字索引）
vol['week'] = date.isocalendar().week.values  #提取 ISO 标准周数，用 .values 转成数组（关键！）
vol['year'] = date.year #提取年份
vol.pivot_table(index=['year', 'week'], columns='stock', values='Volume',aggfunc='sum')
#透视表：按 (year, week) 分组，按 stock 列拆分，求和 volume
```
